```python
#1. Загрузите данные о чемпионах автогонок «Формула 1» с сайта 
#https://en.wikipedia.org/wiki/List_of_Formula_One_World_Drivers%27_Champions 
#(используйте таблицу с указанием всех чемпионов по годам)
# Для этого будет использована библиотека requests для запроса к странице

# URL страницы
url = "https://en.wikipedia.org/wiki/List_of_Formula_One_World_Drivers%27_Champions"
# Установка заголовка User-Agent
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
# Запрос к странице
response = requests.get(url, headers=headers)

# Проверка статуса ответа
if response.status_code == 200:
    # Загрузка всех таблиц с веб-страницы с использованием pandas
    tables = pd.read_html(response.text)
    
    # Вывод количества таблиц и их названий
    print(f"Найдено {len(tables)} таблиц:")
    for i, table in enumerate(tables):
        print(f"Таблица {i}:")
        print(table.head(2))  # Вывод первых нескольких строк таблицы для примера
        print("\n")  
    # Нужная таблица с индексом 2
    champions_table = tables[2]
    # Сохранение таблицы в CSV файл
    champions_table.to_csv('formula_one_drivers_champions.csv', index=False)
    print("Таблица успешно сохранена в файл 'formula_one_drivers_champions.csv'.")
else:
    print(f"Ошибка при доступе к странице: {response.status_code}")
```

```python
# Загрузка данных
data = pd.read_csv('formula_one_drivers_champions.csv')
# Вывод общей информации о датафрейме
print('Информация о загруженных данных:')
print(data.info())
```

```python
# Вывод данных
print(data)
```

```python
#2. Посмотрите результат загрузки, при необходимости удалите лишние строки 
#(подсказка: «подвал» таблицы, то есть последние несколько строк, не нужны).
 
# Удаление первой строки (индекс 0) и последних двух строк (индексы -2 и -1)
data = data.drop(index=[0, data.index[-1], data.index[-2]])
print(data.info())
```

```python
print(data)
```

```python
#3. Проверьте тип данных. Преобразуйте числовые данные (возраст, число поулов, 
#побед, подиумов, быстрейших кругов, очки и т.д.) к числовому типу.

# Переименование столбца 'Constructor' на 'Constructor_Chassis'
data = data.rename(columns={'Constructor': 'Constructor_Chassis'})
# Переименование столбца 'Constructor.1' на 'Constructor_Engine'
data = data.rename(columns={'Constructor.1': 'Constructor_Engine'})

# Изменение типа данных столбцов на int64 и float
data['Age'] = data['Age'].astype('int64')
data['Poles'] = data['Poles'].astype('int64')
data['Wins'] = data['Wins'].astype('int64')
data['Podiums'] = data['Podiums'].astype('int64')
data['Fastest laps'] = data['Fastest laps'].astype('int64')
data['Points'] = data['Points'].astype('float')
data['# of rounds remaining'] = data['# of rounds remaining'].astype('int64')
print(data.info())
```

```python
#1) Средний, минимальный и максимальный возраст гонщиков, становившихся чемпионами.
# Вычисление среднего, минимального и максимального возраста
average_age = data['Age'].mean()
min_age = data['Age'].min()
max_age = data['Age'].max()

# Вывод результатов
print(f"Средний возраст чемпионов: {average_age}")
print(f"Минимальный возраст чемпионов: {min_age}")
print(f"Максимальный возраст чемпионов: {max_age}")
```

```python
#2) Гонщики какой команды выиграли больше всего личных титулов? (Под 
#командой в этом вопросе подразумевается конструктор шасси.) 

data['Count'] = 1 # Добавление нового столбца, заполненного 1, решил количество считать через отдельный столбец
team = data.groupby('Constructor_Chassis')['Count'].sum()

# Вывод результатов
print(f'Команда с наибольшим количеством личных титулов: \n{team[team == team.max()]}')
print(f'\nКоличество выигранных титулов командой:\n{team.sort_values(ascending=False)}') # Вывод отсортировааной таблицы для проверки
```

```python
#3) Имена гонщиков, наибольшее число раз становившихся чемпионами. 

names_driver = data.groupby('Driver')['Count'].sum()
# Вывод результатов
print(f'Имена гонщиков, наибольшее число раз становившихся чемпионами:\n{names_driver[names_driver == names_driver.max()]}')
print(f'\nИмена гонщиков, ставших чемпионами, с колличиством раз:\n{names_driver.sort_values(ascending=False)}') # Вывод отсортировааной таблицы для проверки
```

```python
#4) Кто из гонщиков смог стать чемпионом, выиграв менее 30% гонок в сезоне? 
#(Подсказка: число гонок в сезоне можно определить по столбцу, где указано 
#в какой гонке из скольки был завоеван титул.) 
names_driver2 = data[['Season','Driver', 'Clinched[17]', 'Wins']].copy()

# Извлекаем данные из строки столбца Clinched[17]
# Примечание: в столбце Clinched[17], предполагается, что в записе (Round 6 of 8) количество гонок в сезоне второе число то есть 8 (именно второе число и надо)
# А число выигрышей - это столбец Wins
# Если нужно блыло первое число в строке типа (Round 6 of 8), то 'of (\d+)' нужно заменить на '(\d+)'
names_driver2['Total_Races'] = names_driver2['Clinched[17]'].str.extract('of (\d+)').astype(int)
print(names_driver2.info())
```

```python
print(names_driver2)
```

```python
# Добавление столбца с процентом выигранных гонок
names_driver2['Winning races'] = names_driver2['Wins'] / names_driver2['Total_Races']*100
print(f'Гонщики, которые смогли стать чемпионами, выиграв менее 30% гонок в сезоне:\n{names_driver2[names_driver2["Winning races"] < 30]}')
# Wins - выигранные гонки,  Total_Races - количество гонок в сезоне
```

```python
#5) Рассмотрим гонщиков, которые становились чемпионами хотя бы дважды. 
#Рассмотрим промежутки времени между их «соседними» чемпионствами. 
#Каков максимальный перерыв между двумя последовательными 
#чемпионствами одного и того же гонщика? Например, Нельсон Пике 
#становился чемпионом в 1981, 1983 и 1987 годах. Для него такие перерывы 
#составляют 2 года (с 1981 до 1983) и 4 года (с 1983 до 1987), максимальный 
#из них – 4 года. Вам нужно найти максимум среди всех гонщиков. (В этом 
#задании можно не ограничиваться только функциями библиотек, но и 
#написать свой код на Python.)

# Оставляем в столбце Season только цифры, без 1952[a], будет просто 1952 и меняем тип на int
data['Season'] = data['Season'].str.extract('(\d+)').astype(int)
data.info()
```

```python
data.info()
```

```python
#Группируем по гонщику и считаем количество чемпионств
champions_count = data['Driver'].value_counts()
print(champions_count)
```

```python
# Фильтруем только для гонщиков с двумя и более чемпионствами
multiple_champions = champions_count[champions_count >= 2].index.tolist()
#print(multiple_champions)
filtered_data= data[data['Driver'].isin(multiple_champions)]
print(filtered_data.head(6))
```

```python
# Сортируем по гонщику и сезону
filtered_data = filtered_data.sort_values(by=['Driver', 'Season'])
print(filtered_data.head(6))
```

```python
# Вычисляем разницу между последовательными чемпионствами
filtered_data['Championship_Diff'] = filtered_data.groupby('Driver')['Season'].diff()

print(filtered_data.head(5))
```

```python
# Находим максимальный перерыв между чемпионствами для каждого гонщика
max_breaks = filtered_data.groupby('Driver')['Championship_Diff'].max().reset_index()
print(f'Максимальный перерыв между двумя последовательными чемпионствами одного и того же гонщика для всех гонщиков, котрые побеждали хотя бы 2 раза:\n{max_breaks.head}')
```
